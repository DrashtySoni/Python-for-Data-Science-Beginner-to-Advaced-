# **2. seaborn.jointplot()**
- Draw a plot of two variables with bivariate and univariate graphs.

- This function provides a convenient interface to the JointGrid class, with several canned plot kinds. This is intended to be a fairly lightweight wrapper; if you need more flexibility, you should use JointGrid directly.

## **Syntax :**
**seaborn.jointplot**(*, x=None, y=None, data=None, kind='scatter', color=None, height=6, ratio=5, space=0.2, dropna=False, xlim=None, ylim=None, marginal_ticks=False, joint_kws=None, marginal_kws=None, hue=None, palette=None, hue_order=None, hue_norm=None, **kwargs)

```python
#In the simplest invocation, assign x and y to create a scatterplot (using scatterplot()) with marginal histograms (using histplot()):
sns.jointplot(data=df, x="TotalNumberofEntries", y="CountofNestle")

#Assigning a hue variable will add conditional colors to the scatterplot and draw separate density curves (using kdeplot()) on the marginal axes
sns.jointplot(data=df, x="TotalNumberofEntries", y="CountofNestle", hue="RandomState")

#Set kind="reg" to add a linear regression fit (using regplot()) and univariate KDE curves
sns.jointplot(data=df, x="TotalNumberofEntries", y="CountofNestle", kind="reg")

#There are also two options for bin-based visualization of the joint distribution. The first, with kind="hist", uses histplot() on all of the axes
sns.jointplot(data=df, x="TotalNumberofEntries", y="CountofNestle", kind="hist")

#Alternatively, setting kind="hex" will use matplotlib.axes.Axes.hexbin() to compute a bivariate histogram using hexagonal bins:
sns.jointplot(data=df, x="TotalNumberofEntries", y="CountofNestle", kind="hex")

#Additional keyword arguments can be passed down to the underlying plots
sns.jointplot(
    data=df, x="TotalNumberofEntries", y="CountofNestle",
    marker="+", s=100, marginal_kws=dict(bins=25, fill=False),
)

#Use JointGrid parameters to control the size and layout of the figure
sns.jointplot(data=df, x="TotalNumberofEntries", y="CountofNestle", height=5, ratio=2, marginal_ticks=True)

#To add more layers onto the plot, use the methods on the JointGrid object that jointplot() returns:
g = sns.jointplot(data=df, x="TotalNumberofEntries", y="CountofNestle")
g.plot_joint(sns.kdeplot, color="r", zorder=0, levels=6)
g.plot_marginals(sns.rugplot, color="r", height=-.15, clip_on=False)

# reg kind without NA values
sns.jointplot(x = "TotalNumberofEntries", y = "CountofNestle",kind = "reg", data = df,dropna = True)
```

# **3. seaborn.pairplot()**
- Plot pairwise relationships in a dataset.

- By default, this function will create a grid of Axes such that each numeric
variable in data will by shared across the y-axes across a single row and
the x-axes across a single column. The diagonal plots are treated
differently: a univariate distribution plot is drawn to show the marginal
distribution of the data in each column.

## **Syntax :**
**seaborn.pairplot**(data, *, hue=None, hue_order=None, palette=None, vars=None, x_vars=None, y_vars=None, kind='scatter', diag_kind='auto', markers=None, height=2.5, aspect=1, corner=False, dropna=False, plot_kws=None, diag_kws=None, grid_kws=None, size=None)

The simplest invocation uses scatterplot() for each pairing of the variables and histplot() for the marginal plots along the diagonal

```python
sns.pairplot(df)
```

Assigning a hue variable adds a semantic mapping and changes the default marginal plot to a layered kernel density estimate (KDE):

```python
sns.pairplot(df, hue="RandomState",palette='viridis')
```

It’s possible to force marginal histograms:

```python
sns.pairplot(df, hue="RandomState", diag_kind="hist", palette='gist_rainbow')
```

The kind parameter determines both the diagonal and off-diagonal plotting style. Several options are available, including using kdeplot() to draw KDEs:

```python
sns.pairplot(df, kind="hist",palette='viridis')
```

The markers parameter applies a style mapping on the off-diagonal axes. Currently, it will be redundant with the hue variable

```python
sns.pairplot(df, hue="RandomState", markers=["o", "s", "D"])
```

As with other figure-level functions, the size of the figure is controlled by setting the height of each individual subplot:

```python
sns.pairplot(df, height=5.5)
```

Use vars or x_vars and y_vars to select the variables to plot:

```python
sns.pairplot(
    df,
    x_vars=["TotalNumberofEntries", "CountofNestle", "CountofKellogg"],
    y_vars=["RandomState", "CountofOtherBrands"],
)
```

Set corner=True to plot only the lower triangle:

```python
sns.pairplot(df, corner=True)
```

The plot_kws and diag_kws parameters accept dicts of keyword arguments to customize the off-diagonal and diagonal plots, respectively:

```python
sns.pairplot(
    df,
    plot_kws=dict(marker="+", linewidth=3),
    diag_kws=dict(fill=False,  linewidth=3),
    palette='viridis'
)
```

# **5. seaborn.catplot()**
Figure-level interface for drawing categorical plots onto a FacetGrid.

This function provides access to several axes-level functions that show the relationship between a numerical and one or more categorical variables using one of several visual representations. The kind parameter selects the underlying axes-level function to use:

**Categorical scatterplots:**
- stripplot() (with kind="strip"; the default)
- swarmplot() (with kind="swarm")

**Categorical distribution plots:**
- boxplot() (with kind="box")
- violinplot() (with kind="violin")
- boxenplot() (with kind="boxen")

**Categorical estimate plots:**
- pointplot() (with kind="point")
- barplot() (with kind="bar")
- countplot() (with kind="count")

## **Syntax :**
**seaborn.catplot**(*, x=None, y=None, hue=None, data=None, row=None, col=None, col_wrap=None, estimator=<function mean at 0x7ff320f315e0>, ci=95, n_boot=1000, units=None, seed=None, order=None, hue_order=None, row_order=None, col_order=None, kind='strip', height=5, aspect=1, orient=None, color=None, palette=None, legend=True, legend_out=True, sharex=True, sharey=True, margin_titles=False, facet_kws=None, **kwargs)

```python
#Facet along the columns to show a third categorical variable:
ax = sns.catplot(x="TotalNumberofEntries", y="CountofNestle", hue="RandomState", data=df, col="SplitRatio")
```

```python
#Use a different height and aspect ratio for the facets:
ax = sns.catplot(x="TotalNumberofEntries", y="CountofNestle", hue="RandomState", data=df, col="SplitRatio", height=5, aspect=.8)
```

# **6. seaborn.barplot()**
- Show point estimates and confidence intervals as rectangular bars.

- A bar plot represents an estimate of central tendency for a numeric
variable with the height of each rectangle and provides some indication of
the uncertainty around that estimate using error bars. Bar plots include 0
in the quantitative axis range, and they are a good choice when 0 is a
meaningful value for the quantitative variable, and you want to make
comparisons against it.

- For datasets where 0 is not a meaningful value, a point plot will allow you to focus on differences between levels of one or more categorical variables.
## **Syntax :**
**seaborn.barplot**(*, x=None, y=None, hue=None, data=None, order=None, hue_order=None, estimator=<function mean at 0x7ff320f315e0>, ci=95, n_boot=1000, units=None, seed=None, orient=None, color=None, palette=None, saturation=0.75, errcolor='.26', errwidth=None, capsize=None, dodge=True, ax=None, **kwargs)

```python
sns.set(rc={"figure.figsize": (20, 10)})
#Draw a set of vertical bar plots grouped by a categorical variable:
subplot(4,3,1)
sns.barplot(x="TotalNumberofEntries", y="CountofNestle", data=df,palette='Wistia')

#Draw a set of vertical bars with nested grouping by a two variables:
subplot(4,3,2)
sns.barplot(x="TotalNumberofEntries", y="CountofNestle", hue='RandomState', data=df,palette='viridis')

#Draw a set of horizontal bars:
subplot(4,3,3)
sns.barplot(x="CountofNestle", y="TotalNumberofEntries", hue='RandomState', data=df, orient="h",palette='gist_earth')

#Control bar order by passing an explicit order:
subplot(4,3,4)
sns.barplot(x="TotalNumberofEntries", y="CountofNestle", data=df)

#Use median as the estimate of central tendency:
subplot(4,3,5)
sns.barplot(x="TotalNumberofEntries", y="CountofNestle", data=df, estimator=median,palette='PuBuGn')

#Show the standard error of the mean with the error bars:
subplot(4,3,6)
sns.barplot(x="TotalNumberofEntries", y="CountofNestle", data=df, ci=68,palette='YlGnBu')

#Show standard deviation of observations instead of a confidence interval:
subplot(4,3,7)
sns.barplot(x="TotalNumberofEntries", y="CountofNestle", data=df, ci="sd",palette='Paired')

#Add “caps” to the error bars:
subplot(4,3,8)
sns.barplot(x="TotalNumberofEntries", y="CountofNestle", data=df, capsize=.2, palette='Dark2')

#Use a different color palette for the bars:
subplot(4,3,9)
sns.barplot(x="TotalNumberofEntries", y="CountofNestle", data=df, palette='Spectral')

#Use hue without changing bar position or width:
subplot(4,3,10)
df["SplitRatio"] = df["CountofNestle"].isin([14])
ax = sns.barplot(x="CountofNestle", y="TotalNumberofEntries", hue="RandomState", data=df, dodge=False)

#Plot all ba10s in a single color:
subplot(4,3,11)
ax = sns.barplot(x="TotalNumberofEntries", y="CountofNestle", data=df, color="salmon", saturation=.5)

#Use matplot10ib.axes.Axes.bar() parameters to control the style.
subplot(4,3,12)
ax = sns.barplot(x="TotalNumberofEntries", y="CountofNestle", data=df, linewidth=2.5, facecolor=(1, 1, 1, 0), errcolor=".2", edgecolor=".2")
```

# **7. seaborn.countplot()**

An array or list of vectors.

# **8. seaborn.violinplot()**
- Draw a combination of boxplot and kernel density estimate.

- A violin plot plays a similar role as a box and whisker plot. It shows the distribution of quantitative data across several levels of one (or more) categorical variables such that those distributions can be compared. Unlike a box plot, in which all of the plot components correspond to actual datapoints, the violin plot features a kernel density estimation of the underlying distribution.

- This can be an effective and attractive way to show multiple distributions of data at once, but keep in mind that the estimation procedure is influenced by the sample size, and violins for relatively small samples might look misleadingly smooth.

- Input data can be passed in a variety of formats, including:

 - Vectors of data represented as lists, numpy arrays, or pandas Series objects passed directly to the x, y, and/or hue parameters.

 - A “long-form” DataFrame, in which case the x, y, and hue variables will determine how the data are plotted.

 - A “wide-form” DataFrame, such that each numeric column will be plotted.

 - An array or list of vectors.
 
## **Syntax :**
**seaborn.violinplot**(*, x=None, y=None, hue=None, data=None, order=None, hue_order=None, bw='scott', cut=2, scale='area', scale_hue=True, gridsize=100, width=0.8, inner='box', split=False, dodge=True, orient=None, linewidth=None, color=None, palette=None, saturation=0.75, ax=None, **kwargs)

```python
sns.violinplot(x="TotalNumberofEntries", y="CountofOtherBrands", data=df[df.CountofNestle < 1000], cut=0, scale="width", palette="Set3")
```

- Input data can be passed in a variety of formats, including:

 - A “long-form” DataFrame, in which case the x, y, and hue variables will determine how the data are plotted.

**seaborn.stripplot**(*, x=None, y=None, hue=None, data=None, order=None, hue_order=None, jitter=True, dodge=False, orient=None, color=None, palette=None, size=5, edgecolor='gray', linewidth=0, ax=None, **kwargs)

# **10. seaborn.swarmplot()**
- Draw a categorical scatterplot with non-overlapping points.

- This function is similar to stripplot(), but the points are adjusted (only along the categorical axis) so that they don’t overlap. This gives a better representation of the distribution of values, but it does not scale well to large numbers of observations. This style of plot is sometimes called a **“beeswarm”**.

- A swarm plot can be drawn on its own, but it is also a good complement to a box or violin plot in cases where you want to show all observations along with some representation of the underlying distribution.

- Arranging the points properly requires an accurate transformation between data and point coordinates. This means that non-default axis limits must be set before drawing the plot.

- Input data can be passed in a variety of formats, including:

 - Vectors of data represented as lists, numpy arrays, or pandas Series objects passed directly to the x, y, and/or hue parameters.

 - A “long-form” DataFrame, in which case the x, y, and hue variables will determine how the data are plotted.

 - A “wide-form” DataFrame, such that each numeric column will be plotted.

 - An array or list of vectors.

## **Syntax :**
**seaborn.swarmplot**(*, x=None, y=None, hue=None, data=None, order=None, hue_order=None, dodge=False, orient=None, color=None, palette=None, size=5, edgecolor='gray', linewidth=0, ax=None, **kwargs)

```python
sns.set(rc={"figure.figsize": (30, 20)})
sns.set_theme(style="whitegrid")

#Draw a single horizontal swarm plot:
subplot(2,5,1)
sns.swarmplot(x=df["TotalNumberofEntries"])

#Group the swarms by a categorical variable:
subplot(2,5,2)
sns.swarmplot(x="TotalNumberofEntries", y="CountofOtherBrands", data=df)

#Draw horizontal swarms:
subplot(2,5,3)
sns.swarmplot(x="TotalNumberofEntries", y="CountofOtherBrands", data=df, orient='h')

#Color the points using a second categorical variable:
subplot(2,5,4)
sns.swarmplot(x="TotalNumberofEntries", y="CountofOtherBrands", hue='RandomState', data=df)

#Split each level of the hue variable along the categorical axis:
subplot(2,5,5)
sns.swarmplot(x="TotalNumberofEntries", y="CountofOtherBrands", hue='RandomState', data=df, palette="Set2", dodge=True)

#Control swarm order by passing an explicit order:
subplot(2,5,6)
sns.swarmplot(x="TotalNumberofEntries", y="CountofOtherBrands", data=df, order=['1','2','3'])

#Plot using larger points:
subplot(2,5,7)
sns.swarmplot(x="TotalNumberofEntries", y="CountofOtherBrands", data=df, size=6)

#Draw swarms of observations on top of a box plot:
subplot(2,5,8)
ax = sns.boxplot(x="TotalNumberofEntries", y="CountofOtherBrands", data=df, whis=np.inf)
ax = sns.swarmplot(x="TotalNumberofEntries", y="CountofOtherBrands", data=df, color=".2")

#Draw swarms of observations on top of a violin plot:
subplot(2,5,9)
ax = sns.violinplot(x="TotalNumberofEntries", y="CountofOtherBrands", data=df, inner=None)
ax = sns.swarmplot(x="TotalNumberofEntries", y="CountofOtherBrands", data=df, edgecolor="gray")

#Use catplot() to combine a swarmplot() and a FacetGrid. This allows grouping within additional categorical variables. Using catplot() is safer than using FacetGrid directly, as it ensures synchronization of variable order across facets:
subplot(2,5,10)
g = sns.catplot(x="TotalNumberofEntries", y="CountofOtherBrands",
                 hue="RandomState", col="SplitRatio",
                 data=df, kind="swarm",
                 height=4, aspect=.7);
```

# **11. seaborn.FacetGrid()**
- Multi-plot grid for plotting conditional relationships.

- Initialize the matplotlib figure and FacetGrid object.

- This class maps a dataset onto multiple axes arrayed in a grid of rows and columns that correspond to levels of variables in the dataset. The plots it produces are often called **“lattice”, “trellis”, or “small-multiple” graphics**.

- It can also represent levels of a third variable with the hue parameter, which plots different subsets of data in different colors. This uses color to resolve elements on a third dimension, but only draws subsets on top of each other and will not tailor the hue parameter for the specific visualization the way that axes-level functions that accept hue will.

- The basic workflow is to initialize the FacetGrid object with the dataset and the variables that are used to structure the grid. Then one or more plotting functions can be applied to each subset by calling FacetGrid.map() or FacetGrid.map_dataframe(). Finally, the plot can be tweaked with other methods to do things like change the axis labels, use different ticks, or add a legend. See the detailed code examples below for more information.

## **Syntax :**
**__init__**(self, data, *, row=None, col=None, hue=None, col_wrap=None, sharex=True, sharey=True, height=3, aspect=1, palette=None, row_order=None, col_order=None, hue_order=None, hue_kws=None, dropna=False, legend_out=True, despine=True, margin_titles=False, xlim=None, ylim=None, subplot_kws=None, gridspec_kws=None, size=None)

Calling the constructor requires a long-form data object. This initializes the grid, but doesn’t plot anything on it:

```python
sns.FacetGrid(df)
```

Assign column and/or row variables to add more subplots to the figure:

```python
sns.FacetGrid(df, col="TotalNumberofEntries", row="CountofNestle")
```

To draw a plot on every facet, pass a function and the name of one or more columns in the dataframe to FacetGrid.map():

```python
sns.set(rc={"figure.figsize": (30, 10)})
sns.FacetGrid(df, col="RandomState", row="SplitRatio").map(sns.scatterplot, "TotalNumberofEntries", "CountofNestle")
```

The variable specification in FacetGrid.map() requires a positional argument mapping, but if the function has a data parameter and accepts named variable assignments, you can also use FacetGrid.map_dataframe():

```python
sns.FacetGrid(df, col="RandomState", row="SplitRatio").map_dataframe(sns.histplot, x="TotalNumberofEntries")
```

Notice how the bins have different widths in each facet. A separate plot is drawn on each facet, so if the plotting function derives any parameters from the data, they may not be shared across facets. You can pass additional keyword arguments to synchronize them. But when possible, using a figure-level function like displot() will take care of this bookkeeping for you:

```python
sns.FacetGrid(df, col="RandomState", row="SplitRatio").map_dataframe(sns.histplot, x="TotalNumberofEntries", binwidth=2, binrange=(0, 60))
```

The FacetGrid constructor accepts a hue parameter. Setting this will condition the data on another variable and make multiple plots in different colors. Where possible, label information is tracked so that a single legend can be drawn:

```python
sns.FacetGrid(df, col="CountofOtherBrands", hue="RandomState").map_dataframe(sns.scatterplot, x="TotalNumberofEntries", y="CountofNestle").add_legend()
```

When hue is set on the FacetGrid, however, a separate plot is drawn for each level of the variable. If the plotting function understands hue, it is better to let it handle that logic. But it is important to ensure that each facet will use the same hue mapping.

```python
sns.FacetGrid(df, col="CountofOtherBrands").map_dataframe(sns.scatterplot,  x="TotalNumberofEntries", y="CountofNestle", hue="RandomState").add_legend()
```

The size and shape of the plot is specified at the level of each subplot using the height and aspect parameters:

```python
sns.FacetGrid(df, col="TotalNumberofEntries", height=3.5, aspect=.65).map(sns.histplot, "RandomState")
```

To add horizontal or vertical reference lines on every facet, use FacetGrid.refline():

```python
sns.FacetGrid(df, col="RandomState", margin_titles=True).map_dataframe(sns.scatterplot, x="TotalNumberofEntries", y="CountofNestle").refline(y=df["CountofNestle"].median())
```

- Outliers fall far from the main group of data points. In our plot, after transforming the data set, we have plotted a scatter plot for the first three features.
- *** Syntax ***
> scatter_3d(data_frame=None, x=None, y=None, z=None, color=None, symbol=None, size=None, text=None, hover_name=None, hover_data=None, custom_data=None, error_x=None, error_x_minus=None, error_y=None, error_y_minus=None, error_z=None, error_z_minus=None, animation_frame=None, animation_group=None, category_orders=None, labels=None, size_max=None, color_discrete_sequence=None, color_discrete_map=None, color_continuous_scale=None, range_color=None, color_continuous_midpoint=None, symbol_sequence=None, symbol_map=None, opacity=None, log_x=False, log_y=False, log_z=False, range_x=None, range_y=None, range_z=None, title=None, template=None, width=None, height=None)
