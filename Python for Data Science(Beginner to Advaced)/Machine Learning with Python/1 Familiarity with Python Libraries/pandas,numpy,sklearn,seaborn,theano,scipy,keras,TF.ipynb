# **Theano**
---

### **About the Theano Library**
Theano is a Python library that allows you to define, optimize, and evaluate mathematical expressions involving multi-dimensional arrays efficiently.

Theano has been developed to implement, compile, and evaluate mathematical expressions very efficiently with a strong focus on multidimensional arrays (tensors) which has a NumPy-like syntax (actually, it's built on top of NumPy), and it runs on either CPU or GPU.

### Some of the **features** include:

1. **Automatic Differentiation** – you only have to implement the forward (prediction) part of the model, and Theano will automatically figure out how to calculate the gradients at various points, allowing you to perform gradient descent for model training.
2. **Transparent use of a GPU** – you can write the same code and run it either on CPU or GPU. More specifically, Theano will figure out which parts of the computation should be moved to the GPU.  
3. **Speed and Stability Optimisations** – Theano will internally reorganise and optimise your computations, in order to make them run faster and be more numerically stable. It will also try to compile some operations into C code, in order to speed up the computation. 



>> #### **1.1** Simple Theano Training Example

>>> **1.1.1 Interpretation** - 
The first time the function is called, the output value is still 0.9, because the updates have not been applied yet. But with each consecutive step, the output value becomes closer and closer to the desired target 20.

```python
##### Import Theano Library #####
import theano as tn
 
input = tn.tensor.fvector('input') ### tn variable x to be a vector of 32-bit floats naming as 'input'
target = tn.tensor.fscalar('target') ### Created variable which acts as 'target' value
 
weight = tn.shared(np.asarray([0.2, 0.7]), 'weight') ### asiigned weights to input
output = (input * weight).sum() ### weighted sum formula for actual output
 
cost = tn.tensor.sqr(target - output) # Cost function for training using a simple squared distance
gradients = tn.tensor.grad(cost, [weight]) ### Calculates partial gradients for the parameters that will be updated with respect to the CF.
weight_updated = weight - (0.1 * gradients[0]) ### Using GD, the update rule is to subtract the gradient, multiply by  learning rate

'''Creates a list of updates. 1st param is the variable we want to update, 
and 2nd param is a variable containing the values that we want the first variable to contain after the update.'''
updates = [(weight, weight_updated)]
 
f = tn.function([input, target], output, updates=updates) ## Function passed with i/p and target, output, weight updates
 
for i in range(10):
    output = f([1.0, 1.0], 20.0)
    print (output)
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
from scipy import optimize
#Frequency in terms of Hertz
freq  = 5 
#Sample rate
freq_sample = 50
t = np.linspace(0, 2, 2 * freq_sample, endpoint = False )
a = np.sin(freq  * 2 * np.pi * t)
def function(a):
       return a*2 + 20 * np.sin(a)
plt.plot(a,function(a))
plt.show()

'''If we don’t find a neighbor of global minima, then we need to apply global optimization 
and find global minima function used as basinhopping() which combines local optimizer.'''
optimize.basinhopping(function,0) 
```

### 1.1 Imports

```python
try:
  # %tensorflow_version only exists in Colab.
  %tensorflow_version 2.x
except Exception:
    pass
import tensorflow as tf
from tensorflow.python.keras.utils.vis_utils import plot_model
import pydot
from tensorflow.keras.models import Model
```

### 1.2 Functional API

```python
def build_model_with_functional():
    
    # instantiate the input Tensor
    input_layer = tf.keras.Input(shape=(28, 28))
    
    # stack the layers using the syntax: new_layer()(previous_layer)
    flatten_layer = tf.keras.layers.Flatten()(input_layer)
    first_dense = tf.keras.layers.Dense(128, activation=tf.nn.relu)(flatten_layer)
    output_layer = tf.keras.layers.Dense(10, activation=tf.nn.softmax)(first_dense)
    
    # declare inputs and outputs
    func_model = Model(inputs=input_layer, outputs=output_layer)
    
    return func_model
build_model_with_functional()
```

### 1.3 Build the model and visualize the model graph

```python
# https://pypi.python.org/pypi/pydot
!pip install graphviz && pip install pydot
import pydot
```

```python
from tensorflow.keras.layers import Wrapper
model = build_model_with_functional()

# Plot model graph
plot_model(model, show_shapes=True, show_layer_names=True, to_file='model.png')
```

### 1.4 Training the model

```python
# prepare fashion mnist dataset
mnist = tf.keras.datasets.fashion_mnist
(training_images, training_labels), (test_images, test_labels) = mnist.load_data()
training_images = training_images / 255.0
test_images = test_images / 255.0
```

```python
# configure, train, and evaluate the model
model.compile(optimizer=tf.optimizers.Adam(),
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
```

```python
model.fit(training_images, training_labels, epochs=5)
```

```python
model.evaluate(test_images, test_labels)
```
